## PDB program function.

```python
def PDB_program():
    """
    This function takes no argument.\n
    After calling this function different options
    will be displayed.\n
    A user must read in a file(s) using R option so as to work with
    them for other options.\n
    For more descriptions for each option a user can use
    H option to display options infromation.
    """
    option_menu = ( """
    OPTIONS: 
            READ (R)

            SEARCH (S)

            WRITE (W)

            INFORMATION (I)

            ALIGNMENT (A)

            HELP (H)

            QUIT (Q)
    """)    
    print(option_menu)


    global option
    option = input("Choose an option:") 
    #option_list = ['R','S','W','I','A','H','Q']
    global file_list



    while option.lower() in ('r','s','w','i','a','h','q'):
        if option.lower() in ('r','s','w','i','a','h','q'):

            #if option not in ('R','S','W','I','A','H','Q'):
             #   print("Invalid option \nEnter a valid option")
              #  option_list = ['R','S','W','I','A','H','Q']
                #if option_list:
                    #option = input(":")

            if option.lower() == 'r':
                #global file_list
                #file_path = []
                file_list = []
                import os.path
                global file
                global path
                file = input("Enter the name of the PDB file: ")
                path = "/home/enezer/python-mini-project-enezermjema/data/"
                global filepath
                filepath = path + file
                if os.path.exists(filepath) == True:
                    file_list.append(file)
                    print("File has been successful loaded")
                    #file_path.append(filepath)

                    print(option_menu)
                    option = input("Choose an option:")

                else:
                    print("The file does not exist \nPlease enter a valid name")
                    file = input("Enter the name of the PDB file: ")

            #print(option_menu)
            #option = input("Choose an option:")
            
            
            if option.lower() == 's':
                import re

                print(file_list)
                File = input("choose a file to work with:")
                file_path = path + File

                with open(file_path, "r") as output:
                   # with open("/home/enezer/python-mini-project-enezermjema/data/1dm9.pdb", 'w') as enezer:
                    residues = []
                    list1 = []
                    for line in output:
                        if line.startswith("SEQRES"):
                            linesplit = line.split()
                            if linesplit[2] == "A":
                                residues.append(linesplit[4:])
                    Amino_dict = {'ALA':'A','CYS':'C','ASP':'D','GLU':'E','PCA':'E','PHE':'F','GLY':'G','HIS':'H','ILE':'I','LYS':'K','LEU':'L','MET':'M',
                               'ASN':'N','PRO':'P','GLN':'Q','ARG':'R','SER':'S','THR':'T','VAL':'V','TRP':'W','TYR':'Y','LLP':'K'}

                    single_code  = 'TRIMDYIIYRFLLIFGAPTMALFVLTKAYPFLQNYGYAPTLNIPITGSMDTVYSNSTLAPTREEIFYAFTLTSTLCLYYGAWGP'
                    for lis_list in residues:                                                    
                        for residue in list1:                          
                            list1.append(residue)
                    #global three_letter
                    three_letter = "".join(Amino_dict[x] for x in list1)

                    pattern = re.compile(r'[GYL]A[PFW][TLVGMT]')
                    matches = pattern.finditer(single_code)
                    matches_tup = tuple(matches)
                    print("The number of glycoslation site is %d"%(len(matches_tup)))

                    print(option_menu)
                    option = input("Choose an option:")


            if option.lower() == 'w':
                print("""
                Options:
                        1. Write out coordinate file
                        2. Write out sequence (Fasta format) 
                """)
                write_list = ['1','2']
                choice = input(':')

                if choice == '1':
                    print(file_list)
                    File = input("choose a file to work with:")
                    file_path = path + File
                    new = input("Name for the new Coordinate file:")
                    file_path2 = path + new

                    print("""
                    Sub-options:
                                a. All atom 
                                b. Backbone atoms
                                c. Alpha carbon atoms only
                    """)
                    sub_option = input('Sub-option:')
                    sub_list = ['a','b','c']


                    if sub_option == 'a':
                        with open(file_path, 'r') as amino:
                            with open(file_path2, 'w') as atom:
                                for line in amino:
                                    if line.startswith('ATOM'):
                                        atom.writelines(line)

                        print("File %s has been created successfully"%new)

                        print(option_menu)
                        option = input("Choose an option:")

                    elif sub_option == 'b':
                        with open(file_path, 'r') as amino:
                            with open(file_path2, 'w') as backbone:
                                for line in amino:
                                    if line.startswith("ATOM"):
                                        l = line.split()
                                        atom = l[2]
                                        if atom in ["CA","C","N","O"]:
                                            ene = print(l[2],l[6],l[7],l[8], sep ='\t')
                                            backbone.writelines(ene)

                        print("File %s has been created successfully"%new)

                        print(option_menu)
                        option = input("Choose an option:")

                    elif sub_option == 'c':
                        with open(file_path, 'r') as amino:
                            with open(file_path2, 'w') as alpha:
                                for line in amino:
                                    if line.startswith("ATOM"):
                                        l = line.split()
                                        atom = l[2]
                                        if atom in ["CA"]:
                                            ene = print(l[2],l[6],l[7],l[8], sep ='\t')
                                            alpha.writelines(ene)

                        print("File %s has been created successfully"%new)

                        print(option_menu)
                        option = input("Choose an option:")

                elif choice == '2':
                    print(file_list)
                    File = input("choose a file to work with:")
                    file_path = path + File


                    print("""
                    Sub-options:
                               1. SEQRES Sequence
                               2. Coordinate Sequence
                               3. Alignment Sequence 
                    """)
                    sub_option = input('Sub-option:')
                    if sub_option == '1':
                        new = input("Name for the new SEQRES file:")
                        file_path2 = path + new
                        with open(filepath, 'r') as amino:
                            with open(file_path2, 'w') as seqres:
                                for line in amino:
                                    if line.startswith("SEQRES"):
                                        linesplit = line.split()
                                        if linesplit[2] == "A":
                                            seqres.writelines(line)

                        print("File %s has been created successfully"%new)

                        print(option_menu)
                        option = input("Choose an option:")

                    elif sub_option == '2':
                        new = input("Name for the new COORDINATE file")
                        file_path2 = path + new
                        with open(filepath, 'r') as amino:
                            with open(file_path2, 'w') as coordinate:
                                for x in amino:
                                    if x.startswith("ATOM"):
                                        z = x.split()
                                        coordinate_index = z[6:10]
                                        y = "\t".join(coordinate_index)
                                        j = y+'\n'
                                        coordinate.writelines(j)

                        print("File %s has been created successfully"%new)

                        print(option_menu)
                        option = input("Choose an option:")


            if option.lower() == 'i':
                print(file_list)
                File = input("choose a file to work with:")
                file_path = path + File

                print("""
                SUB-OPTIONS:
                        Display coordinate sequence (1)
                        Display SEQRES sequence (2)
                        Display Alignment sequence (3)
                        Display all non-water ligands in the protein (4)
                """)
                choice = input('Enter a Sub-opton:')
                if choice == '1':
                    with open(filepath, 'r') as amino:
                        for x in amino:
                            if x.startswith("ATOM"):
                                z = x.split()
                                coordinate_index = z[6:10]
                                y = "\t".join(coordinate_index)
                                j = y+'\n'
                                print(j)

                                print(option_menu)
                                option = input("Choose an option:")

                elif choice == '2':
                    print(file_list)
                    File = input("choose a file to work with:")
                    file_path = path + File
                    with open(filepath, 'r') as amino:
                        for line in amino:
                            if line.startswith("SEQRES"):
                                linesplit = line.split()
                                if linesplit[2] == "A":
                                    print(line)

                                    print(option_menu)
                                    option = input("Choose an option:")


            if option.lower() == 'a':
                print(file_list)
                #print(file_path)
                file1 = input("choose a file to work with:")
                file2 = input("choose a file to work with:")
                filepath1 = path + file1
                filepath2 = path + file2
                new = input("Name for the Alignment file:")
                file_path3 = path + new


                from Bio import pairwise2
                from Bio.pairwise2 import format_alignment

                with open(path + file1, "r") as pdb1:
                    with open(path + file2, "r") as pdb2:
                        with open(file_path3, 'w') as pdb3:
                            for line in pdb1:
                                if line.startswith("SEQRES"):
                                    linesplit = line.split()
                                    if linesplit[2] == "A":
                                        global X
                                        X = (line[18:])
                                        #print(X)
                            for line in pdb2:
                                if line.startswith("SEQRES"):
                                    linesplit = line.split()
                                    if linesplit[2] == "A":
                                        global Y
                                        Y = (line[18:])
                                        #print(Y)
                            alignments = pairwise2.align.globalxx(X, Y)
                            for a in alignments:
                                print(format_alignment(*a))
                            stri = ', '.join(map(str, alignments))
                            pdb3.writelines(stri)

                    #for a in alignments:
                     #   aligned = print(format_alignment(*a))
                      #  pdb3.writelines(aligned)


                            print(option_menu)
                            option = input("Choose an option:")

            if option.lower() == 'h':
                print("""

                THE FOLLOWING ARE OPTIONS IN THE PROGRAM WITH THEIR SHORT DESCRIPTION.

                        READ (R) - This option load in a PDB file that is present in the data diretory.

                        SEARCH (S) - Gives a list of PDB files already loaded in the system.\n
                                     It gives the number of glycoslation sites present in the
                                     file choosen and display sequence with glycoslation motifs 
                                     shown in lower case and the rest in upper case.

                        WRITE (W) -  This option write out a file of a given name by the user depending
                                     on the following sub-option chosen;\n
                                              1. Write out coordinate file 
                                                   a. All atom 
                                                   b. Backbone atoms 
                                                   c. Alpha carbon atoms only 
                                              2.Write out sequence (Fasta format) 
                                                   1. SEQRES Sequence
                                                   2. Coordinate Sequence
                                                   3. Alignment Sequence 

                        INFORMATION (I) - This option has sub-options for:\n
                                            1. Displaying coordinate sequence
                                            2. Displaying SEQRES sequence 
                                            3. Displaying Alignment sequence
                                            4. Displaying all non-water ligands in the protein

                        ALIGNMENT (A) - Creates and write to a file an alignment of two PDB sequence 
                                        from two chosen files.\n

                        HELP (H) - This option display all the options in the program with their description.\n

                        QUIT (Q) - Terminate the program.\n

                        THANK YOU...

                        """)

                print(option_menu)
                option = input("Choose an option:")

            if option.lower() == 'q': 
                print(
                """
                PROGRAM TERMINATED\n
                BYE...
                """)
                break


        else:
            print("Invalid option \nEnter a valid option")
            option = input("Choose an option:")
```
